```python
import pandas as pd

# تحميل البيانات (تأكد من اسم الملف الصحيح)
df = pd.read_csv("student_lifestyle_dataset..csv")
df.head()
```

```python
# معلومات عن الأعمدة والبيانات المفقودة
df.info()
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder

# ملء القيم الفارغة بالقيمة الأكثر تكراراً أو الوسيط
df.fillna(df.mode().iloc[0], inplace=True)

# تحويل الأعمدة الفئوية إلى أرقام باستخدام LabelEncoder
encoder = LabelEncoder()
for col in df.select_dtypes(include=['object']).columns:
    df[col] = encoder.fit_transform(df[col])
```

```python
print(df.columns.tolist())
```

```python
target_col = "Grades"

X = df.drop(columns=[target_col, "Student_ID"])  # نستبعد Student_ID لأنه لا يحمل معلومات مفيدة
y = df[target_col]
```

```python
from sklearn.preprocessing import LabelEncoder

# تحويل درجات الحروف إلى أرقام
label_encoder = LabelEncoder()
y = label_encoder.fit_transform(df["Grades"])

# حذف الأعمدة غير المفيدة
X = df.drop(columns=["Grades", "Student_ID"])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import classification_report, accuracy_score

# تقسيم البيانات
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# تدريب النموذج
model = GaussianNB()
model.fit(X_train, y_train)

# التنبؤ والنتائج
y_pred = model.predict(X_test)

# التقييم
print("✅ Accuracy:", accuracy_score(y_test, y_pred))
print("\n📋 Classification Report:\n", classification_report(y_test, y_pred))
```
